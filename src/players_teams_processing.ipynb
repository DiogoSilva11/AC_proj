```python
# Group the DataFrame by 'tmID' and 'year'
grouped = players_teams_data.groupby(['tmID', 'year'])

# Create a dictionary of DataFrames, each representing a unique combination
df_dict = {}
for name, group in grouped:
    tmID, year = name
    sorted_group = group.sort_values(by='minutes', ascending=False)  # Sort and reset index
    df_dict[f'DF_{tmID}_{year}'] = sorted_group.copy() # Subset name = DF_tmID_year

# Access individual DataFrames by their keys in the result dictionary
for key, df_subset in df_dict.items():
    print(f"DataFrame: {key}")
    print(df_subset)
    print()
```

```python

# Assuming you already have grouped your data by 'tmID' and 'year' as 'grouped'

# Define a function to add columns for previous years' stats
def add_previous_years_stats(group):
    if len(group) >= 3:
        tmID, year = group.iloc[0]['tmID'], group.iloc[0]['year']
        for year_offset in range(1, 3):
            previous_year = year - year_offset
            previous_year_subset = players_teams_data[players_teams_data['year'] == previous_year]
            # Merge the previous year's stats into the current group
            group = group.merge(previous_year_subset, on='playerID', how='left', suffixes=('', f'_prev{year_offset}'))
        return group
    else:
        return group

# Apply the function to each group
new_grouped = grouped.apply(add_previous_years_stats)

print(new_grouped)
```

```python
# Access and print the updated DataFrame for a specific team-year combination
desired_team_year = ('CHI', 9)  # Replace with the team and year you want to retrieve

# Filter the DataFrame based on the specified team-year combination

#get a list of all combinations of team and year
team_year_combos = new_grouped[['tmID', 'year']].drop_duplicates().values.tolist()

list_of_dfs = []
columns_to_process = [col for col in new_grouped.columns if (col.endswith('prev1') or col.endswith('prev2')) and not col.startswith('Post') and not col.startswith('stint')]

for year in range(2, 12):
    # Filter data for the specific year
    year_data = new_grouped[new_grouped['year'] == year]

    # Calculate factor
    factor = (2 / 3) - (0.5 * (year_data['assists_prev1'].sum() / year_data['fgMade_prev1'].sum())) / (
                2 * (year_data['fgMade_prev1'].sum() / year_data['ftMade_prev1'].sum()))

    # Calculate VOP
    VOP = year_data['points_prev1'].sum() / (
                year_data['fgAttempted_prev1'].sum() - year_data['oRebounds_prev1'].sum() +
                year_data['turnovers_prev1'].sum() + 0.44 * year_data['ftAttempted_prev1'].sum())

    # Calculate DRB
    DRB = (year_data['dRebounds_prev1'].sum() + year_data['oRebounds_prev1'].sum()) / year_data[
        'rebounds_prev1'].sum()
    
    lg_FT = year_data['ftMade_prev1'].sum()
    lg_PF = year_data['PF_prev1'].sum()
    lg_FTA = year_data['ftAttempted_prev1'].sum()
    
    
for tup in team_year_combos:
    filtered_df = new_grouped[(new_grouped['tmID'] == tup[0]) & (new_grouped['year'] == tup[1])]
    filtered_df = filtered_df.drop(filtered_df.filter(like='Post', axis=1), axis=1)
    filtered_df = filtered_df[filtered_df['stint'] <= 1]
    filtered_df = filtered_df.drop('stint', axis=1)
    filtered_df = filtered_df.drop_duplicates(subset=['tmID', 'year', 'playerID'])
    
    # if columns ending on prev2 are NaN, copy prev1 to prev2, if prev1 and prev2 are NaN, all are 0
    for col in columns_to_process:
        if col.endswith('prev2'):
            filtered_df[col] = filtered_df[col].fillna(filtered_df[col.replace('prev2', 'prev1')])
        elif col.endswith('prev1'):
            filtered_df[col] = filtered_df[col].fillna(0)
        if col.endswith('prev2'):
            filtered_df[col] = filtered_df[col].fillna(0)
    filtered_df['weighted_minutes'] = (filtered_df['minutes_prev1'] * 1 + filtered_df['minutes_prev2'] * 0)

    filtered_df = filtered_df.sort_values(by=('weighted_minutes'), ascending=False)  # Sort and reset index
    filtered_df['weighted_minutes'] = filtered_df['weighted_minutes'].apply(lambda x: x if x >= 48 else 48)

    # Calculate PER for the specific year and update the 'PER' column
    filtered_df['PER'] = (3 * filtered_df['points_prev1'] + (2 / 3) * filtered_df['assists_prev1'] + (
            2 - factor * (filtered_df['assists_prev1'].sum() / filtered_df[filtered_df['tmID'] == tup[0]]['fgMade_prev1'].sum())) * filtered_df['fgMade_prev1'] +
                        (filtered_df['ftMade_prev1'] * 0.5 * (1 + (1 - (filtered_df['assists_prev1'].sum() /
                                                                    filtered_df['fgMade_prev1'].sum())) + (2 / 3) * (
                                                                        filtered_df['assists_prev1'].sum() /
                                                                        filtered_df['fgMade_prev1'].sum()))) - VOP *
                        filtered_df['turnovers_prev1'] - VOP * DRB * (
                                    filtered_df['fgAttempted_prev1'] - filtered_df['fgMade_prev1']) - VOP * 0.44 * (
                                    0.44 + (0.56 * DRB)) * (
                                    filtered_df['ftAttempted_prev1'] - filtered_df['ftMade_prev1']) + VOP * (
                                    1 - DRB) * (
                                    filtered_df['rebounds_prev1'] - filtered_df['oRebounds_prev1']) + VOP * DRB * filtered_df[
                            'oRebounds_prev1'] + VOP * filtered_df['steals_prev1'] + VOP * DRB * filtered_df[
                            'blocks_prev1'] - filtered_df['PF_prev1'] * (
                                    (lg_FT / lg_PF) - 0.44 * (
                                        lg_FTA / lg_PF) * VOP)
                                        ) / filtered_df['weighted_minutes']
    
    #filtered_df = filtered_df.head(10)
    
    list_of_dfs.append(filtered_df)
```

```python
# Concatenate the list of DataFrames into a single DataFrame
concatenated_df = pd.concat(list_of_dfs)

# Select columns that end with 'prev1' or 'prev2'
#prev_columns = [col for col in concatenated_df.columns if col.endswith(('prev1', 'prev2')) or col == 'weighted_minutes' or col == 'height' or col == 'weight']
prev_columns = [col for col in concatenated_df.columns if col.endswith(('prev1', 'prev2')) or col == 'weighted_minutes' or col == 'height' or col == 'weight' or col == 'PER']

for year in range(2, 12):
    # Filter data for the specific year
    year_data = concatenated_df[concatenated_df['year'] == year]

    # Average PER
    avg_PER = year_data['PER'].mean()

    year_data['PER'] = year_data['PER'].multiply(15).div(avg_PER)

    # Update the original DataFrame with the calculated PER for the specific year
    concatenated_df.loc[concatenated_df['year'] == year, 'PER'] = year_data['PER']

# Sum the selected columns for each group
summed_df = concatenated_df[prev_columns]
summed_df = summed_df.drop(summed_df.filter(like='year', axis=1), axis=1)
summed_df = summed_df.drop(summed_df.filter(like='stint', axis=1), axis=1)
summed_df = summed_df.drop(summed_df.filter(like='tmID', axis=1), axis=1)
summed_df = summed_df.drop(summed_df.filter(like='lgID', axis=1), axis=1)
summed_df = summed_df.drop(summed_df.filter(like='dq', axis=1), axis=1)
#summed_df = summed_df.drop(summed_df.filter(like='college', axis=1), axis=1)
#summed_df = summed_df.drop(summed_df.filter(like='Date', axis=1), axis=1)
#summed_df = summed_df.drop(summed_df.filter(like='season', axis=1), axis=1)
#summed_df = summed_df.drop(summed_df.filter(like='pos', axis=1), axis=1)
#summed_df = summed_df.drop(summed_df.filter(like='height_', axis=1), axis=1)
#summed_df = summed_df.drop(summed_df.filter(like='weight_', axis=1), axis=1)

ratings_df = summed_df.groupby(['tmID', 'year']).sum()
ratings_df = ratings_df.drop(ratings_df.filter(like='GP', axis=1), axis=1)
ratings_df = ratings_df.drop(ratings_df.filter(like='GS', axis=1), axis=1)
ratings_df = ratings_df.drop(ratings_df.filter(like='minutes', axis=1), axis=1)
ratings_df = ratings_df.drop(ratings_df.filter(like='Minutes', axis=1), axis=1)

#summed_df = summed_df.drop(summed_df.filter(like='rebounds', axis=1), axis=1)
#summed_df = summed_df.drop(summed_df.filter(like='Attempted', axis=1), axis=1)
#summed_df = summed_df.drop(summed_df.filter(like='Made', axis=1), axis=1)

# List of column names with the suffix prev1
columns_to_divide = [col for col in summed_df.columns if col.endswith('_prev1')]

# Function to calculate the mean of the top n values in a column
def mean_top_n(df, n, column):
    return df.nlargest(n, column)[column].mean()

# Group by 'tmID' and 'year', then calculate the mean of the top 10 'PER' values for each group
mean_per_top_10 = summed_df.groupby(['tmID', 'year']).apply(mean_top_n, n=13, column='PER').reset_index(name='PER')

# Group by 'tmID' and 'year', sum all columns except 'PER', and calculate mean for 'PER'

summed_df = summed_df.drop(summed_df.filter(like='PER', axis=1), axis=1)

summed_df = summed_df.groupby(['tmID', 'year']).sum()

summed_df = summed_df.merge(mean_per_top_10, on=['tmID', 'year'], how='left') 

# Divide columns with the suffix prev1 by column D
summed_df['weighted_minutes'] = summed_df['weighted_minutes'].div(5)
summed_df[columns_to_divide] = summed_df[columns_to_divide].multiply(48).div(summed_df['weighted_minutes'], axis=0)

summed_df = summed_df.drop(summed_df.filter(like='GP', axis=1), axis=1)
summed_df = summed_df.drop(summed_df.filter(like='GS', axis=1), axis=1)
summed_df = summed_df.drop(summed_df.filter(like='minutes', axis=1), axis=1)
summed_df = summed_df.drop(summed_df.filter(like='Minutes', axis=1), axis=1)

# Display the summed DataFrame
print(summed_df)
```
