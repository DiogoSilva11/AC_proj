```python
# Select and rename ingame stats (relevant stats for the prediction)
columns_to_add = ['tmID', 'playoff', 'o_fgm', 'o_fga', 'o_ftm', 'o_fta', 'o_3pm', 'o_3pa', 'o_oreb', 'o_dreb', 'o_reb', 'o_asts', 'o_pf', 'o_stl', 'o_to', 'o_blk', 'o_pts', 'd_fgm', 'd_fga', 'd_ftm', 'd_fta', 'd_3pm', 'd_3pa', 'd_oreb', 'd_dreb', 'd_reb', 'd_asts', 'd_pf', 'd_stl', 'd_to', 'd_blk', 'd_pts']
columns_name_mapping_s1 = {'tmID': 'tmID', 'playoff': 'playoff_s1', 'o_fgm' : 'o_fgm_s1', 'o_fga' : 'o_fga_s1', 'o_ftm' : 'o_ftm_s1', 'o_fta' : 'o_fta_s1', 'o_3pm' : 'o_3pm_s1', 'o_3pa' : 'o_3pa_s1', 'o_oreb' : 'o_oreb_s1', 'o_dreb' : 'o_dreb_s1', 'o_reb' : 'o_reb_s1', 'o_asts' : 'o_asts_s1', 'o_pf' : 'o_pf_s1', 'o_stl' : 'o_stl_s1', 'o_to' : 'o_to_s1', 'o_blk' : 'o_blk_s1', 'o_pts' : 'o_pts_s1', 'd_fgm' : 'd_fgm_s1', 'd_fga' : 'd_fga_s1', 'd_ftm' : 'd_ftm_s1', 'd_fta' : 'd_fta_s1', 'd_3pm' : 'd_3pm_s1', 'd_3pa' : 'd_3pa_s1', 'd_oreb' : 'd_oreb_s1', 'd_dreb' : 'd_dreb_s1', 'd_reb' : 'd_reb_s1', 'd_asts' : 'd_asts_s1', 'd_pf' : 'd_pf_s1', 'd_stl' : 'd_stl_s1', 'd_to' : 'd_to_s1', 'd_blk' : 'd_blk_s1', 'd_pts' : 'd_pts_s1'}
columns_name_mapping_s2 = {'tmID': 'tmID', 'playoff': 'playoff_s2', 'o_fgm' : 'o_fgm_s2', 'o_fga' : 'o_fga_s2', 'o_ftm' : 'o_ftm_s2', 'o_fta' : 'o_fta_s2', 'o_3pm' : 'o_3pm_s2', 'o_3pa' : 'o_3pa_s2', 'o_oreb' : 'o_oreb_s2', 'o_dreb' : 'o_dreb_s2', 'o_reb' : 'o_reb_s2', 'o_asts' : 'o_asts_s2', 'o_pf' : 'o_pf_s2', 'o_stl' : 'o_stl_s2', 'o_to' : 'o_to_s2', 'o_blk' : 'o_blk_s2', 'o_pts' : 'o_pts_s2', 'd_fgm' : 'd_fgm_s2', 'd_fga' : 'd_fga_s2', 'd_ftm' : 'd_ftm_s2', 'd_fta' : 'd_fta_s2', 'd_3pm' : 'd_3pm_s2', 'd_3pa' : 'd_3pa_s2', 'd_oreb' : 'd_oreb_s2', 'd_dreb' : 'd_dreb_s2', 'd_reb' : 'd_reb_s2', 'd_asts' : 'd_asts_s2', 'd_pf' : 'd_pf_s2', 'd_stl' : 'd_stl_s2', 'd_to' : 'd_to_s2', 'd_blk' : 'd_blk_s2', 'd_pts' : 'd_pts_s2'}

# Create df to serve as training - season 3 prediction
season3_predicton_df = pd.DataFrame()
year3_playoff = teams_data.loc[teams_data['year'] == 3, ['tmID', 'playoff']]
season3_predicton_df[['tmID', 'predicted_playoff_qualification']] = year3_playoff.values

# Add penultimate season stats to df
year1_stats = teams_data[teams_data['year'] == 1][columns_to_add]
season3_predicton_df = pd.merge(season3_predicton_df, year1_stats, on='tmID', how='inner')
season3_predicton_df = season3_predicton_df.rename(columns=columns_name_mapping_s1)

# Add last season stats to df
year2_stats = teams_data[teams_data['year'] == 2][columns_to_add]
season3_predicton_df = pd.merge(season3_predicton_df, year2_stats, on='tmID', how='inner')
season3_predicton_df = season3_predicton_df.rename(columns=columns_name_mapping_s2)
print(season3_predicton_df)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance
label_encoder = LabelEncoder()

# Apply label encoding to the 'tmID' column
season3_predicton_df['tmID'] = label_encoder.fit_transform(season3_predicton_df['tmID'])

season3_predicton_df['predicted_playoff_qualification'] = label_encoder.fit_transform(season3_predicton_df['predicted_playoff_qualification'])

season3_predicton_df['playoff_s1'] = label_encoder.fit_transform(season3_predicton_df['playoff_s1'])

season3_predicton_df['playoff_s2'] = label_encoder.fit_transform(season3_predicton_df['playoff_s2'])

print(season3_predicton_df)
```

```python
#Data Modelling 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

X=season3_predicton_df.drop(['predicted_playoff_qualification'],axis=1)
y=season3_predicton_df['predicted_playoff_qualification']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=5)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
classifier1 = DecisionTreeClassifier()
classifier1.fit(X_train,y_train)
y_pred = classifier1.predict(X_test)

cm_decision_trees = confusion_matrix(y_test,y_pred)
print(cm_decision_trees)

labels = np.unique(y_test)
fig, ax = plt.subplots()
im = ax.imshow(cm_decision_trees, interpolation="nearest", cmap=plt.cm.Blues)
ax.figure.colorbar(im, ax=ax)

# Set labels and title
ax.set(xticks=np.arange(cm_decision_trees.shape[1]),
       yticks=np.arange(cm_decision_trees.shape[0]),
       xticklabels=labels, yticklabels=labels,
       xlabel="Predicted label",
       ylabel="True label",
       title="Decision Trees Confusion Matrix")
plt.xticks(rotation=45, ha="right")

# Loop over data dimensions and create text annotations.
for i in range(cm_decision_trees.shape[0]):
    for j in range(cm_decision_trees.shape[1]):
        ax.text(j, i, format(cm_decision_trees[i, j], "d"),
                ha="center", va="center",
                color="white" if cm_decision_trees[i, j] > cm_decision_trees.max() / 2 else "black")

plt.show()

decision_trees_accuracy = accuracy_score(y_test,y_pred)
print(decision_trees_accuracy)
```
