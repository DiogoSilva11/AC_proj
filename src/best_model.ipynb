```python
from pycaret.classification import *

# Setup PyCaret
classification = setup(data = teams_stats, train_size=0.8, target = 'playoff', ignore_features=['tmID', 'year', 'confID'], session_id=123, fold=10, remove_multicollinearity=True, pca_method=True, normalize=True, transformation=True, fix_imbalance=True, use_gpu=True, experiment_name='basketballPlayoffs')

transformed=get_config("dataset_transformed")

print(transformed)


# Create a LightGBM model with specified hyperparameters
lgbm_model = create_model('lightgbm', bagging_fraction=0.9, bagging_freq=0, boosting_type='gbdt',
               class_weight=None, colsample_bytree=1.0, device='gpu',
               feature_fraction=0.8, importance_type='split',
               learning_rate=0.0001, max_depth=-1, min_child_samples=46,
               min_child_weight=0.001, min_split_gain=0.3, n_estimators=290,
               n_jobs=-1, num_leaves=40, objective=None, random_state=123,
               reg_alpha=0.5, reg_lambda=3, subsample=1.0,
               subsample_for_bin=200000, subsample_freq=0)

# Evaluate the model using predict_model on test data
predict_model(lgbm_model, data=season_11)
```
